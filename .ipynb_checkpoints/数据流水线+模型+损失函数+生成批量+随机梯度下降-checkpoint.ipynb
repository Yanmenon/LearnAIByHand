```python
%matplotlib inline
import random
import torch
from d2l import torch as d2l
```

```python
#构造人造数据集，生成包含噪音的样本\
def synthetic_data(w,b,num_examples):
    """生成y=Xw+b+噪音"""
    X=torch.normal(mean=0,std=1,size=(num_examples,len(w)))
    y=torch.matmul(X,w)+b
    y+=torch.normal(0,0.01,y.shape)
    """y.reshape((-1,1))表示把y转换为二维的1列张量"""
    return X,y.reshape((-1,1))
```

```python
true_w=torch.tensor([2,-3.4])
true_b=4.2
#features和lables都是二阶张量
features,lables=synthetic_data(true_w,true_b,1000)
```

```python
print('features:',features[0],"\nlables:",lables[0])
```

```python
#画出数据的分布图
d2l.set_figsize()
d2l.plt.scatter(features[:,1].detach().numpy(),
                lables.detach().numpy(),1);
```

```python
#定义一个data_iter函数，该函数接收批量大小，特征矩阵和标签向量作为输入，生成大小为
#batch_size小批量
def data_iter(batch_size,features,lables):
    num_examples=len(features)
    """为每个特征生成一个索引来指向这个特征"""
    indices=list(range(num_examples))
    """这些样本是随机读取的，不包含顺序"""
    random.shuffle(indices)
    for i in range(0,num_examples,batch_size):
        """返回一批数据对应的索引,indices后面跟着一个切片"""
        batch_indices=torch.tensor(indices[i:min(i+batch_size,num_examples)])
        """yield表示返回值以后机继续执行这个函数"""
        yield features[batch_indices],lables[batch_indices]
```

```python
batch_size=10
for X,y in data_iter(batch_size,features,lables):
    print(X,'\n',y)
    break
```

```python
#定义初始化模型参数,requires_grad=True表示需要计算梯度
w=torch.normal(0,0.01,size=(2,1),requires_grad=True)
b=torch.zeros(1,requires_grad=True)
```

```python
#定义线性回归模型
def linreg(X,w,b):
    return torch.matmul(X,w)+b
```

```python
#定义损失函数
def squared_loss(y_hat,y):
    """均方损失"""
    return (y_hat-y.reshape(y_hat.shape))**2/2
```

```python
#定义优化算法
def sgd(params,lr,batch_size):
    """更新时候不参与梯度的计算"""
    with torch.no_grad():
        """这里的params可以是一个列表，所以可以依次来表示w,b的值"""
        for param in params:
            """因为计算的损失是一个批量样本的总和，所以我们用批量大小（batch_size）
            来归一化步长，这样步长大小就不会取决于我们对批量大小的选择"""
            param-=lr * param.grad/batch_size
            param.grad.zero_()
```

```python
#训练过程
lr=0.001
num_epochs=10
net=linreg
loss=squared_loss
for epoch in range(num_epochs):
    for X,y in data_iter(batch_size,features,lables):
        l=loss(net(X,w,b),y)
        #因为l的形状是（'batch_size',1）而不是一个标量，l的所有元素都会被加入总和中
        l.sum().backward()
        sgd([w,b],lr,batch_size)
    with torch.no_grad():
        train_l=loss(net(features,w,b),lables)
        print(f'epoch{epoch+1},loss{float(train_l.mean()):f}')
            
```
